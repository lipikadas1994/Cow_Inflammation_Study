````r
### 1. Raw read processing using Trimmomatic
#!/bin/bash

# Load Java module (if needed; comment out if already loaded or not using modules)
# module load java

# Define variables
TRIMMO=Trimmomatic-0.39/trimmomatic-0.39.jar
ADAPTERS=Trimmomatic-0.39/Nextera_adapters.fa
OUTDIR=Trimmomatic_trimmed_reads

# Create output directory
mkdir -p $OUTDIR

# Loop through all R1 files
for R1 in *_R1_001.fastq.gz
do
    SAMPLE=$(basename "$R1" _R1_001.fastq.gz)
    R2="${SAMPLE}_R2_001.fastq.gz"

    # Check if both reads exist
    if [[ ! -f "$R1" || ! -f "$R2" ]]; then
        echo "Missing file(s) for sample $SAMPLE. Skipping..."
        continue
    fi

    echo "Processing $SAMPLE..."

    # Run Trimmomatic with HEADCROP:20 and 36 threads
    java -jar "$TRIMMO" PE -threads 36 \
        "$R1" "$R2" \
        "$OUTDIR/${SAMPLE}_R1_paired.fastq.gz" "$OUTDIR/${SAMPLE}_R1_unpaired.fastq.gz" \
        "$OUTDIR/${SAMPLE}_R2_paired.fastq.gz" "$OUTDIR/${SAMPLE}_R2_unpaired.fastq.gz" \
        ILLUMINACLIP:"$ADAPTERS":2:30:10 \
        LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 HEADCROP:20 MINLEN:50 \
        2> "$OUTDIR/${SAMPLE}_trimmomatic.log"

done

#### Copy the trimmed .paired .fastq.gz reads to DADA2_inputs folder
````

````r
##### Network Analysis at Genus level
library(phyloseq)
library(igraph)
library(ggraph)
library(tidygraph)
library(dplyr)
library(tibble)

# Extract and format taxonomic information
tax_df <- as.data.frame(tax_table(ps_genus))
tax_df$Genus <- ifelse(is.na(tax_df$Genus) | tax_df$Genus == "", "Unclassified", as.character(tax_df$Genus))
tax_df$Phylum <- ifelse(is.na(tax_df$Phylum) | tax_df$Phylum == "", "Unknown", as.character(tax_df$Phylum))

# Assign genus names as taxa names
taxa_names(ps_genus) <- make.unique(tax_df$Genus)

# Prepare OTU matrix (samples as rows, taxa as columns)
otu_mat <- as(otu_table(ps_genus), "matrix")
if (taxa_are_rows(ps_genus)) {
  otu_mat <- t(otu_mat)
}

# Calculate Spearman correlation matrix
cor_mat <- cor(otu_mat, method = "spearman", use = "pairwise.complete.obs")
cor_mat[is.na(cor_mat)] <- 0

# Keep strong positive correlations only
threshold <- 0.6
cor_mat[cor_mat < threshold] <- 0
diag(cor_mat) <- 0  # Remove self-loops

# Create igraph object from correlation matrix
net <- graph_from_adjacency_matrix(cor_mat, weighted = TRUE, mode = "undirected", diag = FALSE)

# Add taxonomic data to nodes
V(net)$Genus <- names(V(net))
V(net)$Phylum <- tax_df[match(V(net)$Genus, rownames(tax_df)), "Phylum"]

# Compute degree and assign node size
V(net)$degree <- degree(net)
V(net)$size <- scales::rescale(V(net)$degree, to = c(3, 10))  # rescale for better visual range

# Plot network
ggraph(net, layout = "fr") +
  geom_edge_link(aes(width = weight), color = "gray70", alpha = 0.5) +
  geom_node_point(aes(color = Phylum, size = size)) +
  geom_node_text(aes(label = Genus), repel = TRUE, size = 3, color = "black") +
  scale_edge_width(range = c(0.2, 2)) +
  theme_void() +
  theme(legend.position = "right") +
  labs(title = "Genus-Level Microbiome Correlation Network", color = "Phylum", size = "Degree")
````

````r
##### Without Singleton
# Keep only the largest connected component
components <- components(net)
largest_component_nodes <- V(net)[components$membership == which.max(components$csize)]
net_filtered <- induced_subgraph(net, vids = largest_component_nodes)
# Plot network with only connected genera
ggraph(net_filtered, layout = "fr") +
  geom_edge_link(aes(width = weight), color = "gray70", alpha = 0.5) +
  geom_node_point(aes(color = Phylum, size = size)) +
  geom_node_text(aes(label = Genus), repel = TRUE, size = 3, color = "black") +
  scale_edge_width(range = c(0.2, 2)) +
  theme_void() +
  theme(legend.position = "right") +
  labs(title = "Genus-Level Microbiome Correlation Network (Connected Component Only)",
       color = "Phylum", size = "Degree")
````
